# Plotting demo


This demo will show how to plot the top n marker genes found in each cluster, and how to visualize pathways annotated for cluster.

```python
from descartes_rpa.io.load import load_data_with_pathways
from descartes_rpa import get_pathways_for_group
```

### Load the AnnData structure

```python
adata = load_data_with_pathways("../data/output/Thymus")
get_pathways_for_group(adata)
```

### Import the plotting functions

```python
from descartes_rpa.pl import marker_genes
```

### Plot top 5 marker genes for clusters

This plot will be saved as a PDF file, being able to change its name, plot type and output directory.

```python
marker_genes(adata, name="Thymus.pdf", n_genes=5)
```

```python
!ls dotplot*
```

### Different plotting format
Possible formats: violin, stacked_volion, heatmap, dotplot (default), matrixplot and tracksplot.

```python
marker_genes(adata, name="Thymus.pdf", plot_format="matrixplot", n_genes=5)
```

```python
!ls matrixplot*
```

### Plotting only 2 genes

```python
marker_genes(adata, name="Thymus.pdf", n_genes=2)
```

## Pathways visualization tool

You can also, inside a Jupyter Notebook, get nice visualizations of pathways annotated to each cluster.

```python
from descartes_rpa.pl import pathways
```

```python
# Checking which clusters exist
adata.uns["pathways"].keys()
```

```python
antigen = pathways(adata, "Vascular endothelial cells")
antigen
```

```python
thymocytes = pathways(adata, "Thymocytes")
thymocytes
```

Plotting shared pathways between all clusters. By default, it shows shared pathways between all clusters.

```python
from descartes_rpa.pl import shared_pathways
```

```python
shared_pathways(adata)
```

```python
!ls shared_pathways.png
```

Seeing only shared pathways between Thymocytes and Stromal Cells. You can use the "shared_pathways" function to see two or more shared pathways between clusters.

```python
shared_pathways(adata, clusters=["Thymocytes", "Stromal cells"], file_name="thymocytes_stromal.png")
```

```python
!ls thymocytes_stromal.png
```
